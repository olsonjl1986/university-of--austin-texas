#### Create a function univ_analyis(df) :Loop through each column and create univariate graphs for each numeric column. 
- 1) the for loop loops through each attribute in the data by using the two indexes
and searches by each item in the column =. within the loop it will iterate over each column and execute to the two if statements
- 2) the if statements will constrain the for loop to only numeric data types and string data types. 
i.e. int, float, ..etc
    - If statement 1: Constrain the data to be plotted to only the numeric data 
        for each iteration it will graph a distribution plot and a histogram for each numeric data type in the dataframe.
    - elif statement 1: Constrain the data to only string type datatypes and plot a countplot for each string datatype   
- 3) The for loop and graphing declarations will be ran through the function univ_analysis 
which has an input of a dataframe.



```python
### Univariate Data Analysis
simp_list = ["Product",["Gender"],["MaritalStatus"]]
def univ_analysis(CardioData):
    
    for cardio, attributes in CardioData.items():
        if pd.api.types.is_numeric_dtype(attributes):
            
            sns.set()
           
            fig, ax = plt.subplots()
            
            
            fig.set_size_inches(10,10)
            
            sns.histplot(x=attributes, data= CardioData)
            
            plt.show()
            sns.boxplot(x=attributes, data= CardioData)
            plt.show()
             
        
        elif pd.api.types.is_string_dtype(attributes):
            sns.set()
           
            fig, ax = plt.subplots()
                        
            fig.set_size_inches(10,10)
            
            
           
            sns.countplot(x=attributes, data= CardioData);       
            plt.show()   
       
            
            

        
```

```python
univ_analysis(CardioData)
```

### MultiVariate Analysis


```python
# define function to perform multivariate analysis and write each graph to file.

# define list of columns to iterate over 
simp_list= ["Miles","Usage","Income","Education"]
list_stringdtype = ["Product","MaritalStatus","Gender"]
def multi_var(CardioData):
    
    # define correlation matrix and heat map to look at correlation between each variable
    
   CardioData.corr()
   plt.figure(figsize=(10,10))
   sns.heatmap(CardioData.corr(), annot=True, linewidths=.5, fmt= '.1f', center = 1 )  # heatmap
   plt.show()
   #return CardioData 
   
    # create barplots for data for all important variable to do multivariate analysis by using the x as a categorical variable
    # and the y as numerical
    # from the correlation heatmap these are the relationships that have > .6 corr factor   

    # plot relationship between Income vs education 
        
   sns.barplot(x=CardioData["Education"],y= CardioData["Income"],data= CardioData, hue = "MaritalStatus");
   plt.show()
   sns.barplot(x=CardioData["Education"],y= CardioData["Income"],data= CardioData, hue = "Product");
   plt.show()
   sns.barplot(x=CardioData["Education"],y= CardioData["Income"],data= CardioData, hue = "Gender");
   plt.show()
   
   # Plot relationship between Usage vs fitness 

   sns.barplot(x=CardioData["Fitness"],y= CardioData["Usage"],data= CardioData, hue = "Gender");
   plt.show()
   sns.barplot(x=CardioData["Fitness"],y= CardioData["Usage"],data= CardioData, hue = "Product");
   plt.show()
   
   sns.barplot(x=CardioData["Fitness"],y= CardioData["Usage"],data= CardioData, hue = "MaritalStatus");
   plt.show()

   # Plot the relationship Miles vs Usage 
   sns.barplot(x=CardioData["Usage"],y= CardioData["Miles"],data= CardioData, hue = "Product");
   plt.show()
   sns.barplot(x=CardioData["Usage"],y= CardioData["Miles"],data= CardioData, hue = "MaritalStatus");
   plt.show()
   sns.barplot(x=CardioData["Usage"],y= CardioData["Miles"],data= CardioData, hue= "Gender");
   plt.show()
    
    # Plot the relationship between Miles vs Fitness
   sns.barplot(x=CardioData["Fitness"],y= CardioData["Miles"],data= CardioData, hue= "Gender");
   plt.show() 
   sns.barplot(x=CardioData["Fitness"],y=CardioData["Miles"], data=CardioData, hue = "Product") 
   plt.show()
   sns.barplot(x=CardioData["Fitness"],y=CardioData["Miles"], data=CardioData, hue = "MaritalStatus")
   plt.show()
   
```
